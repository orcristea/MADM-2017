### Step 1. Import the necessary libraries

```python
# Preparamos el entorno para poder empezar a trabajar. Importamos las librerías básicas
# El "ipython magic" que permite ver los plots en notebooks
%matplotlib inline

# pandas
import pandas as pd

# numpy
import numpy as np

# matplotlib.pyplot
import matplotlib.pyplot as plt

# seaborn
import seaborn as sns
```

### Step 2. Import the dataset from this US_Baby_Names_right.csv 

### Step 3. Assign it to a variable called baby_names.

```python
baby_names = pd.read_csv('data/US_Baby_Names_right.csv')
```

### Step 4. See the first 10 entries

```python
baby_names.head(10)
```

### Step 5. Delete the column 'Unnamed: 0' and 'Id'

```python
# el inplace sirve para no tener que hacer una reasignacion 
# https://stackoverflow.com/questions/13411544/delete-column-from-pandas-dataframe-using-python-del
baby_names.drop('Unnamed: 0', axis=1, inplace=True)
baby_names.drop('Id', axis=1, inplace=True)
```

```python
baby_names.head(10)
```

### Step 6. Is there more male or female names in the dataset?

```python
baby_names['Gender'].value_counts('F')
# el porcentaje de mujeres es mas alto, un 54%
```

### Step 7. Group the dataset by name and assign to names

```python
names = baby_names.groupby("Name").sum()

names.head()
```

### Step 8. How many different names exist in the dataset?

```python
len(names)
# 17632 diferentes nombres
```

### Step 9. What is the name with most occurrences?

```python
names.sort_values("Count", ascending = 0).head()
#Jacob con 242874 repeticiones
```

### Step 10. How many different names have the least occurrences?

```python
#names.Count.min()
len(names[names.Count == names.Count.min()])
```

### Step 11. What is the median name occurrence?

```python
names[names.Count == names.Count.median()]
```

### Step 12. What is the standard deviation of names?

```python
names.Count.std()
```

### Step 13. Get a summary with the mean, min, max, std and quartiles.

```python
names.describe()
```
